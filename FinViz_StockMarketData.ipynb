```python
# Add the dependencies.
import pandas as pd
import os

# Files to load
overview_to_load = os.path.join("Resources", "overview.csv")
performance_to_load = os.path.join("Resources", "performance.csv")


# Read the school data file and store it in a Pandas DataFrame.
overview_df = pd.read_csv(overview_to_load)
performance_df1 = pd.read_csv(performance_to_load)


performance_df1.sample(5)

        
```

```python
new_column_order = ["Ticker", "Performance (Week)","Performance (YTD)", "Volatility (Month)","Average Volume","Performance (Half Year)"]
performance_df= performance_df1[new_column_order]
performance_df.head(5)
```

```python
overview_df.sample(5)
```

```python
stocks_df = pd.merge (overview_df,performance_df,on=["Ticker","Ticker"], how = 'right')
stocks_df.sample(5)
```

```python
total_companies =stocks_df["Company"].count()
print(f" There are {total_companies} in the US stock market")
```

```python
sectors = len(stocks_df["Sector"].unique())
print(f" There are {sectors} sectors in the US stock market")
```

```python
industries = len(stocks_df["Industry"].unique())
print(f" There are {industries} industries in the US stock market")
```

```python
total_market_cap = stocks_df["Market Cap"].sum()

print(f" The total Market cap of the US Stock market is {total_market_cap} usd")
```

```python
Big_market_cap=stocks_df[stocks_df["Market Cap"]>= 100000]
Big_market_cap.sample(5)
```

```python
Big_market_cap_count = Big_market_cap["Company"].count()
print(f" There are {Big_market_cap_count} companies with a market cap over $100M USA today")
```

```python
high_market_cap_and_low_pe = stocks_df[(stocks_df["Market Cap"] >= 100000) & (stocks_df["P/E"] <= 50)]
high_market_cap_and_low_pe.sample(10)               
```

```python
high_market_cap_and_low_pe_count = high_market_cap_and_low_pe["Company"].count()
high_market_cap_and_low_pe_count
print(f" There are {high_market_cap_and_low_pe_count} companies with a market cap over $100M USA and a P/E lower than 50 today")
```

```python
mean_highMC_lowPE=high_market_cap_and_low_pe["Average Volume"].mean()
mean_highMC_lowPE
```

```python
stocks_summary_df = pd.DataFrame(
    [{"Total companies":total_companies,
    "Sectors":sectors,
    "Industries":industries,
    "Total Market Cap":total_market_cap,
    "Companies MarketCap > $100M":Big_market_cap_count,
    "Companies MarketCap > $100M & PE<50":high_market_cap_and_low_pe_count,
    "Average dalily Volume ":mean_highMC_lowPE,},])

stocks_summary_df
```
